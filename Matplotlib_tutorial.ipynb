 #### Implementing two plots in one graph

```python
x = np.arange(1,6)
```

```python
y1 = 2*x
```

```python
y2 = 3*x
```

```python
plt.plot(x,y1, color="g", linestyle=":",linewidth=2)
plt.plot(x,y2, color="yellow", linestyle="--",linewidth=2)
plt.title("graph of x and y")
plt.xlabel("x value")
plt.ylabel("y value")
plt.grid(True)
```

#### Two plots in same screen

```python
# so we can place multiple subplots in a row
# Suppose we want to make 3 subplots in a row

y3 = 3*x
```

```python
plt.subplot(1,3,1)
plt.plot(x,y1, color="g", linestyle="--", linewidth=1)

plt.subplot(1,3,2)
plt.plot(x,y2, color="b", linestyle="--", linewidth=2)
plt.plot(x,x**2, color="y", linestyle="--", linewidth=1)

plt.subplot(1,3,3)
plt.plot(x,y3, color="r", linestyle="--", linewidth=1)

plt.title("graph of x and y")
plt.xlabel("x value")
plt.ylabel("y value")
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1, color="g", linestyle="--", linewidth=1)
plt.show()
plt.subplot(2,1,2)
plt.plot(x,y2, color="b", linestyle="--", linewidth=1)
plt.show()
```

## Bar plot

Marks of students in a bar graph
name of students in a list and their marks in other list.
or you can make a dictionary object and then extract name and marks as list

```python
student ={"Harshit": 98, "Rohit Sharma": 68, "Sneha": 74}
```

```python
student.keys()
```

```python
names = list(student.keys())
```

```python
values = list(student.values())
```

```python
plt.bar(names, values)
plt.grid(True)
plt.title("distribution of marks")
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

```python
plt.barh(names, values, color="g")
plt.show()
```

## Scatter Plot
#### X--> contains some values of coordinates
#### Y--> Contsains some values of coordinates

plt.scatter

```python
x=[1,2,3,4,5]
y=[40,20,10,50,30]
```

```python
plt.scatter(x,y)
plt.show()
plt.grid
```

```python
plt.scatter(x,y, marker="*", color="r", s=100)
plt.show()
```

```python
# Two plots in a same graph

z=[78,5,43,6,76]
plt.scatter(x,z,color="r", marker="*", s=200)
plt.show()
```

```python
## Subplots 

plt.subplot(1,2,1)
plt.scatter(x,y, color="r", marker="*", s=200)

plt.subplot(1,2,2)
plt.scatter(y,z)
```

## Pie Chart

```python
fruit =['Apple', "guava", 'orange']
count = [23,65,18]
```

```python
plt.pie(count, labels=fruit)
plt.show()
```

#### Percentage in pie chart

```python
plt.pie(count,labels=fruit, colors=['yellow','red','blue'])
plt.show()
```

```python
# DoughNut Chart

plt.pie(count, labels=fruit, radius=2)
plt.pie([1],radius=1, colors=['w'])
plt.show()
```
